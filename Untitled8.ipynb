```python
#RGB Image
import matplotlib.pyplot as plt
import cv2
```

```python
cim = plt.imread(r'Sunflowers.jpg')
```

```python
#View color image 
plt.imshow(cim)
```

```python
#cim = cv2.imread(r'Sunflowers.jpg') #Lists in BGR ordering in OpenCV...will change photo colors.
#To fix this
#cim = cim[:,:,::-1]

#Define each channel
R = cim[:,:,0]
G = cim[:,:,1]
B = cim[:,:,2]
```

```python
plt.figure(1)

plt.subplot(231)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Red')

plt.subplot(235)
plt.imshow(G, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap='gray') #cmap gray - each individual channel looks like grayscale
plt.xticks([])
plt.yticks([])
plt.title('Blue')

plt.show()
```

```python
#Creating red, white and black patches on pictures.

R[100:400, 100:400] = 255 #Position of each square.
G[100:400, 100:400] = 0 
B[100:400, 100:400] = 0

cim[:,:,0] = R
cim[:,:,1] = G
cim[:,:,2] = B

plt.figure(1)

plt.subplot(231)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Red')

plt.subplot(235)
plt.imshow(G, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap='gray') #cmap gray - each individual channel looks like grayscale
plt.xticks([])
plt.yticks([])
plt.title('Blue')

plt.show()
```

```python
#Create RGB and BGR
import numpy as np
import pandas as pd
rgb = np.zeros((100,150,3),dtype='uint8') #RGB
rgb[:,0:50,0] = 255 #Red
rgb[:,50:100,1] = 255 #Green
rgb[:,100:150,2] = 255 #Blue
plt.imshow(rgb)
```

```python
#In CV2, the ordering is BGR
cv2.imshow('RGB',rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
